# EDA

## Overall score, histogram

```python
fig, ax = plt.subplots(1, 1, figsize = (10, 6))
sns.histplot(x='overall', data=df, hue='d_div1_league', binwidth=1, multiple='dodge', kde=True, palette='Blues', ax=ax)
ax.set(xlabel='Overall score', title='Distribution of overall score, 2015 - 2020')
```

## Overall score vs. player tags/traits, violin plots

```python
# player tags
d_tags = [x for x in df.columns if x.startswith('d_tag')]

fig, axs = plt.subplots(5, 4, figsize = (15, 15))
for i, ax in enumerate(axs.ravel()):
    if i >= len(d_tags):
        ax.set_visible(False)
    else:       
        sns.violinplot(data=df, x=d_tags[i], y='overall', palette='Blues', ax=ax)
        ax.set(xlabel=d_tags[i], ylabel='Overall score', title=d_tags[i])
fig.suptitle('Overall score vs. player tags, 2015-2020')
plt.tight_layout(rect=[0, 0.03, 1, 0.99])
```

```python
# player traits
d_traits = [x for x in df.columns if x.startswith('d_trait')]

fig, axs = plt.subplots(8, 5, figsize = (15, 20))
for i, ax in enumerate(axs.ravel()):
    if i >= len(d_traits):
        ax.set_visible(False)
    else:       
        sns.violinplot(data=df, x=d_traits[i], y='overall', palette='Blues', ax=ax)
        ax.set(xlabel=d_traits[i], ylabel='Overall score', title=d_traits[i])
fig.suptitle('Overall score vs. player traits, 2015-2020')
plt.tight_layout(rect=[0, 0.03, 1, 0.99])
```

## Overall score vs. other variables, barplots

```python
var_list = ['preferred_foot','international_reputation','weak_foot','skill_moves','d_multiple_position']

fig, axs = plt.subplots(2, 3, figsize = (12, 8))
for i, ax in enumerate(axs.ravel()):
    if i >= len(var_list):
        ax.set_visible(False)
    else:       
        sns.violinplot(data=df, x=var_list[i], y='overall', palette='Blues', ax=ax)
        ax.set(xlabel=var_list[i], ylabel='Overall score', title=var_list[i])
fig.suptitle('Overall score vs. other indicators, 2015-2020')
plt.tight_layout(rect=[0, 0.03, 1, 0.99])
```

```python
df['age_cut'] = pd.cut(df['age'], bins=8, precision=0)
fig, ax = plt.subplots(1, 1, figsize = (10, 6))
sns.barplot(x='age_cut', y='overall', data=df, color='skyblue')
ax.set(xlabel='Age', ylabel='Overall score', title='Distribution of overall score, 2015 - 2020')
```

## Positions, correlation matrix

```python
d_position = [x for x in df.columns if x.startswith('d_pos')]
df_pos = df[df.columns.intersection(d_position)]
df_pos.columns = [x.replace('d_pos_', '') for x in df_pos.columns]

fig, ax = plt.subplots(1, 1, figsize = (12, 8))
sns.heatmap(df_pos.corr(), annot=True, fmt='.1f', linewidths=.5, cmap='Blues', ax=ax)
ax.set(title='Correlation matrix, positions')
plt.show()
```

```python
# correlation matrix, positions vs. player traits
pos_tag = d_position + d_tags
df_pos_tag = df[df.columns.intersection(pos_tag)]
pos_tag_cor = df_pos_tag.corr()
pos_tag_cor = pos_tag_cor[d_tags].filter(like='d_pos', axis=0)
pos_tag_cor.index = [x.replace('d_pos_', '') for x in pos_tag_cor.index]
pos_tag_cor.columns = [x.replace('d_tag_', '') for x in pos_tag_cor.columns]

fig, ax = plt.subplots(1, 1, figsize = (12, 8))
sns.heatmap(pos_tag_cor, annot=True, fmt='.1f', linewidths=.5, cmap='Blues', ax=ax)
ax.set(title='Correlation matrix, positions vs. player tags')
plt.show()
```

```python
# correlation matrix, positions vs. player traits
pos_trait = d_position + d_traits
df_pos_trait = df[df.columns.intersection(pos_trait)]
pos_trait_cor = df_pos_trait.corr()
pos_trait_cor = pos_trait_cor[d_traits].filter(like='d_pos', axis=0)
pos_trait_cor.index = [x.replace('d_pos_', '') for x in pos_trait_cor.index]
pos_trait_cor.columns = [x.replace('d_trait_', '') for x in pos_trait_cor.columns]

fig, ax = plt.subplots(1, 1, figsize = (20, 8))
sns.heatmap(pos_trait_cor, annot=True, fmt='.1f', linewidths=.5, cmap='Blues', ax=ax)
ax.set(title='Correlation matrix, positions vs. player traits')
plt.show()
```

## Overall Score by Player 

Distribution of overall rating of the players:
* Distribution of the overall rating of the players is pretty much normal, but annual wage and the value market have distributions skewed to the right. Then, they won't be perfect predictors to the talent of the player because they could respond to other characteristics of the player like how popular they are (how many shirts they sell, if it's an iconic player for the club etc).
* When ploting the scatter plots of Annual Wage and Market Value of the players versus their overall score, we can detect that some of them are overvalued or overpaid since a few difference in their overall score is associated with a much greater differece in the compensation they received or their market value. Again, this differences could be due to something besides skills like popularity.

```python
df_20 = df[df['year'] != 2020]
fig, ax = plt.subplots(1, 1, figsize = (10, 6))
sns.kdeplot(df_20['overall'], ax=ax)
ax.set(xlabel='Overall Rating', title='Distribution of Overall Rating of Players, 2020')
```
