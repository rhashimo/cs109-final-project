## Overall Score by League

* Looking at the overall Score by league, there is such variability as in nationality. 
* Moreover, when looking at the talent of the players against the annual wage they receive and the market value of the players (size of the marker), it seems that there are overvalued leagues. Again, it could be related to other features not contained in the database, but if we only care about talent of players, some leagues are over valued. 

```python
number_league = df_20.groupby(by="league_name").mean().sort_values(by='overall', ascending=False)
number_league['overall'].head(20)
```

```python
coef_var_lea = number_league['overall'].std() / number_league['overall'].mean()
print(f'{coef_var_lea:.2f}')
```
